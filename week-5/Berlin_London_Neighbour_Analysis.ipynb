<p>6.4.7.1 Let's explore the Venues and venues categories on Berlin neighbourhood</p>

```python
def get_near_by_venues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, CLIENT_SECRET, VERSION, lat, lng, radius, LIMIT)
            
        #     # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
          
          # return only relevant information for each nearby venue
        venues_list.append([(name, lat, lng, 
                               v['venue']['name'], v['venue']['location']['lat'], v['venue']['location']['lng'],
                               v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue in venues_list for item in venue])
    nearby_venues.columns = ['Neighbourhood','Neighbourhood Latitude', 'Neighbourhood Longitude', 
                               'Venue', 'Venue Latitude', 'Venue Longitude', 'Venue Category']
      
    return nearby_venues
 
```

<h3>Perform the onehot encoding by using venue categories as the features sets</h3>


<p>Mostly,our task is to classifying them according to the venues categories they have so we find all unique feature venue categories as feature for input model so we are going use pandas dummies function for onehot encoding</p>

```python
# onehot encoding by the python get_dummied function
Berlin_onehot = pd.get_dummies(Berlin_venues[['Venue Category']], prefix= "", prefix_sep= " ")
Berlin_onehot['Neighbourhood']=Berlin_venues['Neighbourhood']
Berlin_onehot.head(5)
```

<p>This give the frequency of common categories venues in the each neighbourhoods</p>

```python
Berlin_group=Berlin_onehot.groupby(["Neighbourhood"]).mean().reset_index()
Berlin_group.shape
Berlin_group.head()
```

<p>6.4.9 Let's ranking the venues in each neighbourhood according to there ferquecy </p>

```python
def return_most_common_venues(row, num_top_venues):
    row = row.iloc[1:]
    row_sorted = row.sort_values(ascending=False)
    
    return row_sorted.index.values[0:num_top_venues]
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighbourhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))
```

<p>making the dataframe by ranking the most common venues categories for the each neighbourhoods</p>

```python
Berlin_venues_sorted = pd.DataFrame(columns=columns)
Berlin_venues_sorted['Neighbourhood'] = Berlin_group['Neighbourhood']

for ind in np.arange(Berlin_group.shape[0]):
    Berlin_venues_sorted.iloc[ind, 1:] = return_most_common_venues(Berlin_group.iloc[ind, :], num_top_venues)

Berlin_venues_sorted.head()
```

<p>Here we are using k=5 so our feature datasets are classiyfing into total 5 class and class label as 0,1,2,3,4 and we just fixed random_state for center assign as 0.<p>

```python
from sklearn.cluster import KMeans
```

<p>we first clustering the data into 5 class and making the merged of two dataframe which having column as geoposition values,ranking of the venues on the each neighbourhoods</p>

```python
# I'm considering number of clusters as 5.
k = 5
kmeans = KMeans(n_clusters=k, random_state=0)
Berlin_model_input =Berlin_group.drop('Neighbourhood', axis = 1)
# fit the model by the feature selected data
Berlin_cluster=kmeans.fit(Berlin_model_input)
# assign the label for each neighbour
Berlin_venues_sorted['Cluster_label']=Berlin_cluster.labels_
Berlin_final = N_df.copy()
# merge top venues_sorted with toronto_data
Berlin_final = Berlin_final.join(Berlin_venues_sorted.set_index('Neighbourhood'), on='Neighbourhood')

Berlin_final.head() # check the last columns!
```
